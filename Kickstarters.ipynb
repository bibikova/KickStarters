```python
country_list = ['GB', 'US', 'CA', 'NO', 'AU', 'IT', 'DE', 'IE', 'ES', 'N,"0', 'MX','SE', 'FR', 'NL', 'NZ', 'CH', 'AT', 'BE', 'DK', 'HK', 'LU']
```

```python
df_c2 = df_c1[df_c1['country'].isin(country_list)]
```

```python
df_c2.info()
```

```python
# now we can drop all those Unnamed columns cause they are empty
df_c2 = df_c2.drop(columns = ['Unnamed:13', 'Unnamed:14', 'Unnamed:15', 'Unnamed:16'], axis = 1)
```

```python
df_c2.sample()
```

```python
#let's take care of time data
df_c2.deadline[0]
```

```python
df_c2['deadline'] = df_c2['deadline'].str.split(' ').str.get(0)
df_c2['launched'] = df_c2['launched'].str.split(' ').str.get(0)
```

```python
df_c2.sample()
```

```python
df_c2['deadline'] = pd.to_datetime(df_c2['deadline'])
df_c2['launched'] = pd.to_datetime(df_c2['launched'])
```

```python
df_c2.info()
```

```python
df_c2.sample()
```

```python
#and turn columns with numbers to integers 
df_c2[["goal", "pledged", "backers", "usdpledged"]] = df_c2[["goal", "pledged", "backers", "usdpledged"]].apply(pd.to_numeric)
```
